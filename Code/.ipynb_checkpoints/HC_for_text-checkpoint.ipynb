```r
#simple text processing functions

truncate_text = function(raw_text, max_length) {
  # truncate tail of raw_text to have max_length words (at most)
  paste(unlist(strsplit(raw_text, " "))[1:max_length], sep = " ", collapse = " ")
}

sample_words = function(text, n) {
 #returns n words from text by randomly sample them without replacements
  paste(sample(unlist(strsplit(text, " ")), n, replace = FALSE), sep = " ", collapse = " ")
}

num_of_words = function(raw_text) {
  #count number of words in raw_text
  length(unlist(strsplit(raw_text, " ")))
}

count_words = function(raw_text, 
                      ignore_list = c(""),
                      min.counts = 0) {
  # return the number of appearance of each word in raw_text
  # pre-processing removes punctuation and words from the ignore_list
  words <- tolower(raw_text)
  words <- gsub("[[:punct:]]", " ", words)
  words <- gsub('[0-9]+', " ", words)
  words <- unlist(strsplit(words, " "))
  words <- words[!words == ""]
  words <- words[!words == " "]
  words <- words[!words == "\n"]
#  words <- wordStem(words)
  
  words <- words[!is.element(words, ignore_list)]
  tble <- sort(table(words))
  df <- as.data.frame(tble)
  df <- filter(df, Freq >= min.counts)
  df
}

reduce_counts = function(counts, word_list) {
  # reduce the table of counts to words in word_list only
  # (returns a list of counts_
    
  c <- rep(0, length(word_list))
  
  #df <- counts[counts$words %in% word_list,]
  for(i in seq_along(word_list)) {
    fr = counts$Freq[counts$words == word_list[i]]
    if (length(fr) > 0) {
      c[i] <- fr
    }
  }
  c
}

get_word_frequency = function(text, words) {
    #get frequency of usage of words in list 'words'
        c0 = word_count(text)
        freq = c0 / sum(c0)
        ferq_common_words = reduce_counts(freq, words)
}
```

```r
#auxiliarry functions for comparing two or more distinct texts

compare_text = function(text1, text2, compare_by, alpha = 1, min.counts = 3, alt = "two.sided") {
    #auxiliary function to compare two texts (text1, text2) by a list of words (compare_by)
  l1 = number_of_words(text1)
  l2 = number_of_words(text2)
  
  #truncate longer text to number of words of shorter text
  if(l1 > l2) {  
    #paper1 <- truncate_text(paper1, l2)
    text1 <- sample_words(text1, l2)
  }
  if(l2 > l1) {
    #paper2 <- truncate_text(paper2, l1)
    text2 <- sample_words(text2, l1)
  }
  
  #count words appearing in compare_by
  c1 = reduce_counts(word_count(text1), compare_by)
  c2 = reduce_counts(word_count(text2), compare_by)
  
  pv = p.values.binom_half(c1, c2, min.counts = min.counts, alt = alt)
  #additional 5 p-values based on exact tests
  pv_db = p.value.double(k1,k2)
  pv = c(pv_db, pv)
      for (m in c(1,2,3,4) ) {
      pv.n = p.value.ncount(c1,c2, n=m)
      pv = c(pv, pv.n)}
  hc.vals(pv, alpha = alpha, interp = TRUE)
}

join_papers = function(corpus, range){
  joint = NULL
  for(i in range)
  {
    joint = c(joint, corpus[[i]])
  }
  joint
}
```

<H1>Extract a list of distinguishing words (features) from all available data</H1>
(we use the undisputed Federalists papers that contain enough written matrial by Hamilton, and the additional book: "An examination of the British doctrine... " by Madison. 

```r
#feature exrtraction using all Ham and Mad texts


# this list contains "contextual" words that are not useful for distinguishin between the authors 
ignore_list = c('govern', 'trade', 
                'enemi','govern', 'legislatur','xii', 'z',
                'x', 'v', 'xi', 'xiii', 'vi', 'vii', 'ii',
                'iii', 'ix', 'war','senat', 'constitut', 'treati',
                'government', 'senate', 'president', 'jurisdiction',
               'treaty', 'legislature', 'france', 'state', 's', 'laws','union', 'national', 'power',
                'constitution','constitution', 'security','nations', 'british', 'states',
               'commerce', 'courts', 'enemies', 'britain', 'citizens','citizen','public', 'system','treaties')

corpus1 = ham.all
corpus2 = c(mad.all, as.character(mad.external))

l1 = number_of_words(corpus1)
l2 = number_of_words(corpus2)

EQUALIZE = TRUE
#truncate longer corups to number of words of shorter corpus
if(EQUALIZE) {
    if(l1 > l2) {  
    #text1 <- truncate_text(text1, l2)
    corpus1 <- sample_words(corpus1, l2)
    }
    if(l2 > l1) {
    #text2 <- truncate_text(text2, l1)
    corpus2 <- sample_words(corpus2, l1)
}}

#count words
c1 = as.data.frame(word_count(corpus1, min.counts = 5, ignore_list = ignore_list))
c2 = as.data.frame(word_count(corpus2, min.counts = 5, ignore_list = ignore_list))
df = merge(c1,c2,by = 'words')

#compute p-values and HC w.r.t. H0 : B(c1+c2,1/2) 
pv = p.values.binom_half(df[['Freq.x']], df[['Freq.y']], min.counts = 80, alt = "two.sided")

hc = hc.vals(pv, alpha = 0.4, interp = FALSE)
print(hc$hc.star)

#extract list of distinguishing words
avail_idcs = which(!is.na(pv))[hc$p.sorted_idx[1:hc$i.max.star]]
feat = df$words
dist_data = data_frame('word' = feat[avail_idcs],
                       'p.value' = pv[avail_idcs],
                       'Ham' = df[['Freq.x']][avail_idcs], 'Mad' = df[['Freq.y']][avail_idcs])
dist_words = feat[avail_idcs]

print("list of distinguishing words (features) and their p-value under B(n = sum.counts,p = 1/2):")
print(dist_data)
```

<H1> Compare all disputed texts using estimated frequencies of each author</H1>

```r
words = as.character(dist_words) #from previous cell

#get typical usage frequencies of distcriminating words by each author
rHam = reduce_counts(word_count(ham.all), word_list = words)
nH = sum(word_count(ham.all, min.counts = 2)$Freq)
rHam = rHam / nH

rMad = reduce_counts(word_count(mad.all), word_list = words)
nM = sum(word_count(mad.all, min.counts = 2)$Freq)
rMad = rMad / nM

# count number of appearences of each word in 'words' in disputed texts:
counts = reduce_counts(word_count(disputed.all, ignore_list = ignore_list), word_list = words);
N = sum(word_count(disputed.all, min.counts = 2)$Freq)

#compute p-values with respect to word frequencies of each author:
pv = p.values.poisson(x = counts, N = N, r = rMad, alt = "two.sided"); 
#Note: Mosteller & Wallace used Beta distribution instead of Poisson to better fit the tails of usage frequencies
hc1 = hc.vals(pv, alpha = 0.75);
MD = hc1$hc.star;
print(sprintf("MD = %f",MD))

pv = p.values.poisson(x = counts, N = N, r = rHam, alt = "two.sided");
hc2 = hc.vals(pv, alpha = 0.75);
HD = hc2$hc.star;
print(sprintf("HD = %f",HD))

dist_data = data_frame('word' = words, 'Mad expected' = N*rMad, 'Ham expected' = N*rHam, 'actual' = counts, 'Mad pval' = hc1$p, 
                       'Ham pval' = hc2$p)
print(dist_data)
#"MD = 6.188325"
#"HD = 7.211285" 

plot(hc1$uu,hc1$z)
title("Mad Z-score")

plot(hc2$uu,hc2$z)
title("Ham Z-score")
```

<H1> Compare all disputed texts using HC</H1>

```r
compare_texts = function(text1, text2, compare_by, n.Monte = 1, max_size = 1e20, min.counts = 3, alpha = 1) {
    # compute HC of p-values with respect to Binom(n,1/2) of counts collected from text1 and text2 
    # since length(text1) != length(text2) we sample words from each text and repeat the experiment n.Monte times
    # the number of words we sample is the minimum of 'max_size' and the size of the texts. 
    # words appearing fewer than 'min.count' in the sample are ignored
    # HC only consider the smallest 'alpha' fractions of p-values 
    size = min(number_of_words(text1), number_of_words(text2), max_size)
    res = rep(0,n.Monte)
    for (i in (1:n.Monte)){
        #count words appearing in compare_by
        c1 = reduce_counts(word_count(sample_words(text1, size)), compare_by)
        c2 = reduce_counts(word_count(sample_words(text2, size)), compare_by)
        pv = p.values.binom_half(c1, c2, min.counts = min.counts, alt = "two.sided")
        #additional 5 p-values from exact tests
        pv_db = p.value.double(c1,c2)
        pv = c(pv_db, pv)
              for (m in c(0,1,2,3,4) ) {
              pv.n = p.value.ncount(c1,c2, n=m)
              pv = c(pv, pv.n)}

        hc <- hc.vals(pv, alpha = alpha)
        res[i] <- hc$hc.star
    }
    res
}
```

```r
# compare against each other (sample from corups and run HC two-sample test)
words = as.character(dist_words) #from previous cell

disputed.all = join_papers(fed.papers, disputed)
joint.all = join_papers(fed.papers, joint)


n.Monte = 30
al = 0.45

size = number_of_words(disputed.all)
smp_HD = compare_texts(ham.all, disputed.all, words, max_size = size ,n.Monte = n.Monte, alpha = al)
#smp_DH = compare_texts(disputed.all, ham.all, words, max_size = size ,n.Monte = n.Monte, alpha = al)
smp_HH = compare_texts(ham.half1, ham.half2, words, max_size = size ,n.Monte = n.Monte, alpha = al)

smp_MD = compare_texts(mad.all, disputed.all, words, max_size = size ,n.Monte = n.Monte, alpha = al)
#smp_DM = compare_texts(disputed.all, mad.all, words, max_size = size ,n.Monte = n.Monte, alpha = al)
smp_MM = compare_texts(mad.half1, mad.half2, words, max_size = size ,n.Monte = n.Monte, alpha = al)

smp_MH = compare_texts(mad.all, ham.all, words, max_size = size ,n.Monte = n.Monte, alpha = al)
smp_HM = compare_texts(ham.all, mad.all, words, max_size = size ,n.Monte = n.Monte, alpha = al)

smp_DD = compare_texts(disputed.half1, disputed.half2, words, max_size = size ,n.Monte = n.Monte, alpha = al)

#size = number_of_words(joint.all)
#smp_JD = compare_texts(joint.all, disputed.all, words, max_size = size,  n.Monte = n.Monte, alpha = al)

print(paste("HH = ",mean(smp_HH)))
#print(paste("HD = ",mean(smp_HD)))
print(paste("DH = ",mean(smp_DH)))

print(paste("MM = ",mean(smp_MM)))
print(paste("MD = ",mean(smp_MD)))
#print(paste("DM = ",mean(smp_DM)))
```

```r
words = as.character(dist_words) #from previous cell

rHam = reduce_counts(word_count(ham.all), word_list = words)
nH = sum(word_count(ham.all, min.counts = 2)$Freq)
rHam = rHam / nH

rMad = reduce_counts(word_count(mad.all), word_list = words)
nM = sum(word_count(mad.all, min.counts = 2)$Freq)
rMad = rMad / nM

disputed.all = join_papers(fed.papers,disputed)
nD = sum(word_count(disputed.all, min.counts = 2)$Freq)

print(paste("# words in Hamilton papers = ", nH))
print(paste("# words in Madison  papers = ", nM))
print(paste("# words in Disputed papers = ", nD))

compare_list = function(idcs, p.null, compare_by, min.counts = 0, alpha = 1){
  #auxiliary function to compare word counts in papers versus p.null  
  #use HC vs null for each paper whose index appears in idcs
  res = rep(0,length(idcs));
  for(i in seq_along(idcs)) 
  { 
    text = fed.papers[[idcs[i]]];
    counts = reduce_counts(word_count(text), word_list = compare_by);
    N = sum(word_count(text, min.counts = 2)$Freq)
    pv = p.values.poisson(x = counts, N = N, r = p.null, min.counts = min.counts, alt = "two.sided");
    hc = hc.vals(pv, alpha = alpha, interp = TRUE);
    res[i] = hc$hc.star;
  }
  res
}

MM = compare_list(mad.known, rMad, min.counts = 0, alpha = 0.4, compare_by = words)
HH = compare_list(ham.known, rHam, min.counts = 0, alpha = 0.4, compare_by = words)
HM = compare_list(mad.known, rHam, min.counts = 0, alpha = 0.4, compare_by = words)
MH = compare_list(ham.known, rMad, min.counts = 0, alpha = 0.4, compare_by = words)
HD = compare_list(disputed, rHam, min.counts = 0, alpha = 0.4, compare_by = words)
MD = compare_list(disputed, rMad, min.counts = 0, alpha = 0.4, compare_by = words)
```

 <H1>Compare each pair of papers according to a list of words</H1>

```r
words = as.character(dist_words) #from previous cells

compare_groups = function(group1, group2, compare_by, n.Monte = 1) {
  #auxiliary function to compare word counts of each paper in group1 with all papers in group2 
  # groupi is a group of incdices to use with fed.papers
  #'compare_by' is a list of words to compare by
  n1 = length(group1)
  n2 = length(group2)
  hc.mat = matrix(data = 0, nrow = n1, ncol = n2)
  num_of_comp = 0
  
  for (i in seq_along(group1)) {
    for (j in seq_along(group2)) {
      if(group1[i] != group2[j])  {
        num_of_comp = num_of_comp + 1
        paper1 <- fed.papers[[group1[i] ]]
        paper2 <- fed.papers[[group2[j] ]]
        
        res = compare_texts(paper1, paper2, compare_by = compare_by,
                           alpha = 0.45, min.counts = 3, n.Monte = n.Monte)
        hc.mat[i,j] <- mean(res)
      }
    }
  }
  mean = sum(hc.mat) / num_of_comp
  hc.mat
}

HH = compare_groups(ham.known, ham.known, 
                    compare_by =  words, n.Monte = 5 )
MM = compare_groups(mad.known, mad.known, 
                    compare_by = words, n.Monte = 5)
HM = compare_groups(ham.known, mad.known, 
                    compare_by = words, n.Monte = 5)
MH = compare_groups(mad.known, ham.known, 
                    compare_by = words, n.Monte = 5)
HD = compare_groups(ham.known, disputed,
                    compare_by = words, n.Monte = 5)
MD = compare_groups(mad.known, disputed,
                  compare_by = words, n.Monte = 5)

JD = compare_groups(joint, disputed,
                  compare_by = words, n.Monte = 5)

DJ = compare_groups(disputed, joint,
                  compare_by = words, n.Monte = 5)

JJ = compare_groups(joint, joint,
                  compare_by = words, n.Monte = 5)

MM = MM[upper.tri(MM, diag = FALSE) | lower.tri(MM, diag = FALSE)] #extract diagonals
HH = HH[upper.tri(HH, diag = FALSE) | lower.tri(HH, diag = FALSE)]
JJ = JJ[upper.tri(JJ, diag = FALSE) | lower.tri(JJ, diag = FALSE)]

qqplot(x = MM, y = MH)
abline(0,1)
```
